```
import pandas as pd
file= 'MotherJones_MassShooting_Clean.xlsx'
xl=pd.ExcelFile(file)
df1=xl.parse('5ae6f6d7-6df1-4887-9ded-35c84f3')
df1.head() #importing first dataframe
```

```
file2= 'SPLC_HateGroups_Clean.xlsx'
xl=pd.ExcelFile(file2)
df2=xl.parse('26422446-e4f1-471f-a912-c8c94b9')
df2.head() #importing second data frame
```

```
count = df1['State'].value_counts()
print(count)
count2= df2['State'].value_counts()
print(count2)
#Counting the instances of mass murders and hate groups by state
```

```
frames = [count,count2]



new_dataframe= pd.concat(frames, axis=1, join='inner', join_axes=None, ignore_index=False,
          keys=None, levels=None, names=None, verify_integrity=False,
          copy=True)
print(new_dataframe)

for col in new_dataframe.columns:
  print(col) #combining the counts into one dataset
```

```
new_dataframe.columns= ['Count_Mass_Murders', 'Count_Hate_Groups']
new_dataframe #updated column headers
```

```
import matplotlib.pyplot as plt
import numpy as np
ax=new_dataframe.plot.scatter (x='Count_Mass_Murders', y='Count_Hate_Groups')
#created scatter plot
```

```
import seaborn as sns; sns.set(color_codes=True)

ax1 = sns.regplot(x="Count_Mass_Murders", y="Count_Hate_Groups", data=new_dataframe)
#created simple regression - slight correlation
```
